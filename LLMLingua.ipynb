```python
contexts = """The Industrial Revolution stands as a defining epoch in human history, marking a fundamental shift from agrarian societies to mechanized, industrialized economies. It emerged in the late 18th century in Britain and swiftly disseminated throughout Europe and the United States. This era witnessed the advent of groundbreaking innovations, notably the steam engine, which catalyzed unprecedented advancements in manufacturing. Initially revolutionizing textile production, these technologies rapidly extended their influence to diverse sectors like iron manufacturing and transportation, igniting a transformative wave across industries.

Central to this period was the profound societal reconfiguration sparked by urbanization. As rural populations migrated to burgeoning urban centers in pursuit of factory employment, cities burgeoned, becoming hubs of industrial activity. This migration wrought both economic metamorphosis and social stratification. While it facilitated the rise of a burgeoning middle class, it simultaneously exacerbated wealth disparities, fostering a burgeoning divide between the affluent and the impoverished.

The Industrial Revolution wasn't solely a narrative of industrial and economic growth; it also bore witness to nascent social reforms and profound shifts in labor dynamics. Labor movements gained momentum, advocating vehemently for improved working conditions and equitable treatment of workers. Simultaneously, the plight of child labor was cast into the spotlight, leading to the implementation of pivotal child labor laws aimed at safeguarding the welfare of the young workforce.

The epochal implications of the Industrial Revolution continue to reverberate through contemporary societies, molding the foundations of modern economies, technological progress, and societal structures on a global scale."""

```

**Question 1**

```python
question = "Question: What characterized the Industrial Revolution's inception?\nAnswer:"
reference = "Transition from agrarian to mechanized economies in late 18th century Britain."
```

```python
# The response from original prompt, using GPT-4-32k
import json
prompt = "\n\n".join([contexts, question])

message = [
    {"role": "user", "content": prompt},
]

request_data = {
    "messages": message,
    "max_tokens": 100,
    "temperature": 0,
    "top_p": 1,
    "n": 1,
    "stream": False,
}
response = openai.ChatCompletion.create(
    deployment_id = "gptturbo",
    **request_data,
)
print(json.dumps(response, indent=4))
```

```python
# Setup LLMLingua
from llmlingua import PromptCompressor
llm_lingua = PromptCompressor()
```

```python
# 200 Compression
compressed_prompt = llm_lingua.compress_prompt(
    contexts.split("\n"),
    instruction="",
    question=question,
    target_token=200,
    condition_compare=True,
    condition_in_question='after',
    rank_method='longllmlingua',
    use_sentence_level_filter=False,
    context_budget="+100",
    dynamic_context_compression_ratio=0.4, # enable dynamic_context_compression_ratio
    reorder_context="sort"
)
message = [
    {"role": "user", "content": compressed_prompt["compressed_prompt"]},
]

request_data = {
    "messages": message,
    "max_tokens": 100,
    "temperature": 0,
    "top_p": 1,
    "n": 1,
    "stream": False,
}
response = openai.ChatCompletion.create(
    deployment_id = "gptturbo",
    **request_data,
)

print(json.dumps(compressed_prompt, indent=4))
print("Response:", response)
```
